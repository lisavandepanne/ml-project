```python
from implementations import *
import datetime
```

```python
y = np.array([1, 4, 5, 10])
print(y.shape)

tx = np.array([[1, 1],
               [1, 3],
               [1, 6],
               [1,10]])
print(tx.shape)

w0 = np.array([0, 0])
print(w0.shape)
```

### Checking MSE_GD

```python
start_time = datetime.datetime.now()
w_opt, loss_opt = mean_squared_error_gd(y, tx, w0, 10000, 0.01)
end_time = datetime.datetime.now()

print(w_opt)
print(w_opt.shape)

# Print result
exection_time = (end_time - start_time).total_seconds()
print("GD: execution time={t:.3f} seconds".format(t=exection_time))
```

### Checking least squares
